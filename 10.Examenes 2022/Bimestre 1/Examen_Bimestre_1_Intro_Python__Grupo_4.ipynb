## Limpieza de datos

````
#Elimino registros con rating con menos de 80 ocurrencias
ratings=df_netflix["rating"].value_counts()
ratings = ratings[ratings <=80]
df_netflix = df_netflix[~df_netflix['rating'].isin(ratings.index)]

#Remplazo de blancos por np.NaN
df_netflix['country']=(df_netflix['country'].replace(r'^\s*$', np.nan, regex=True))
df_netflix
````

## Transformación de datos

````
#Transformamos formato de fecha a formato Pandas, para así poder trabajar con los métodos relacionados
df_netflix['date_added']=pd.to_datetime(df_netflix['date_added'])  
#df_netflix["date_added"]

df_netflix['year'] = df_netflix["date_added"].dt.year
df_netflix["year"] = df_netflix["year"].fillna(0).astype(int)
````

Al tener ordenados los datos de títulos argentinos, ahora podemos proceder a la carga de 10 nuevos títulos

````
#Funcion para cargar contenido al DF NEtflix
def agregar_contenido(tipo,titulo,director,reparto,pais,fecha,anio,clasi,duracion,genero,descripcion,dataframe):
  lista = [tipo,titulo,director,reparto,pais,fecha,anio,clasi,duracion,genero,descripcion]
  df_prueba = pd.DataFrame(lista)
  df_prueba = df_prueba.transpose()
  df_prueba.columns = ['type', 'title', 'director', 'cast', 'country', 'date_added','release_year', 'rating', 'duration', 'listed_in', 'description']
  dataframe = dataframe.append(df_prueba, ignore_index = False)
  return dataframe   
````

## Conclusión

En conclusion, la plataforma Netflix cuenta con un contenido total de 8667 películas y series de las cuales un 69.22% corresponde a la primer categoría y un 30.78% a la segunda. Su rating es mayormente TV-MA lo que indica que su contenido es principalmente apuntado a un público adulto, mientras que en menor medida son menores los porcentajes de contenidos aptos para menores de hasta 14 y 16 años.

Continuando con el contenido de la plataforma, podemos ver como su pico de carga fue en el año 2019, inicalmente la hipotesis del grupo sobre este particular análisis estuvo basada en la teoría de que la mayoría de carga de contenido podría haber sido durante los años 2020 y 2021 debido a la pandemia y la mayor predisposición de la gente a consumir plataformas de streaming por encontrarse en sus hogares más tiempo, pero claramente la industria fue afectada por los confinamientos, muchos proyectos se detuvieron, fueron pospuestos o cancelados, lo que resultó en una disminución en la producción y estreno de nuevos contenidos.

En cuanto a los pincipales productores, se pudo observar como la gran mayoría proviene de Estados Unidos, una clara y conocida potencia de la industria cinematográfica y en segundo lugar, se encuentra India, quien además de haber desarrollado una gran industria conocida como "Bollywood", también cuenta con una de las mayores audiencias y consumidores de contenido cinematográfico nacional por su gran densidad poblacional.

Por último, con el fin de poder analizar un poco más el contenido ofrecido por Netflix, se tomó como comparación su principal competencia, la plataforma Amazon Prime.
Al tomar esta plataforma como punto de comparación pudimos ver algunas similitudes como que ambas tienen mayormente películas a series, y encontramos las principales difernecias en la cantidad total de contenido siendo la suma de peliculas y series de netflix un total de 8667 y un total de 9668 por parte de Amazon.

Otra gran diferencia es la clasificación del rating de Amazon Prime, quien en su mayoría tiene contenido TV +13, lo que indica una priorización de un público general, seguido en menor medida por contenido +16 y +18.
