En primer lugar analizamos las variables numéricas a fin de determinar si presentan una distribución normal o no

```
# Función para graficoar histograma
def histo (variable, nombre):
  # Graficamos el histograma
  plt.hist(variable, density=True, bins = 20)
  plt.xlabel(nombre)

  # Calculamos promedio y DS del vector 
  mu = variable.mean()
  sigma = variable.std()
  # Calciulamos los ejes x y aplicamos la función de DN para y
  x = np.linspace(0, max(variable), len(variable))
  y = sqrt(1/(2*pi*sigma**2))*exp(-(x-mu)**2/(2*sigma**2))
  # Graficamos la curva como 2º capa
  plt.plot(x, y, color = 'red')

  plt.show()
```

```
# Función para graficoar Curva de densidad entre la teórica y la Muestral
def normal_teorica(variable, nombre):
  fig, ax = plt.subplots()

  sns.kdeplot(x = variable, label = 'Muestra')

  x = np.linspace(-2, max(variable)+0.5, len(variable))
  mu = np.mean(variable)
  sd = np.std(variable)
  y = stats.expon.pdf(x, mu, sd)
  ax.plot(x, y, label = 'Teorica')

  ax.legend(loc = 'upper right')
  plt.title(f'Distribución Muestral y Teorica de {nombre}')
  plt.show()
```

Por las evidencias de la evaluación gráfico y test de normalidad la variable `PRCP` no tendría una distribución normal motivo por el cual se implementarán test NO paramétricos para su análisis.
