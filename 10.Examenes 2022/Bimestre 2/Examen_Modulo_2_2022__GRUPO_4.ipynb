Viajar en avión es uno de los transportes más seguros. Sin embargo, cuando ocurren accidentes aéreos, los resultados suelen ser catastróficos, su tasa de supervivencia suele ser muy baja y forma parte de uno de los principales miedos entre la gente.

Incluso en las mejores situaciones de entrenamiento puestas en práctica tanto por los pilotos, como por los tripulantes del vuelo, al momento de un accidente real, se requiere de un accionar preciso. Ya que, según los profesionales, volar un avión es fácil, lo difícil es mantener el control durante situaciones extremas.

Por un lado, podemos tener en cuenta que a la hora de analizar un accidente aéreo, según el Global Aviation Safety Plan (GASP) existen tres tipos principales actualmente:

- El vuelo controlado contra terreno: la colisión contra un obstáculo sin que 
haya una pérdida de control.

- Las incursiones y excursiones en pista: se producen cuando un avión entra en una pista en la que no está autorizado, o le autorizan para despegar donde va a aterrizar otro avión.

- La pérdida de control en vuelo: resulta en el mayor número de fatalidades en aviación comercial. Puede deberse a fallos en los equipos, falta de adiestramientos correctos (falla humana) o falta de monitoreo, entre otras cosas.

Por otro lado, tenemos en cuenta también lo expuesto por la Asociación de Transporte Aéreo Internacional (IATA), que la mayoría de los accidentes aéreos son provocados por errores humanos, que en muchos casos tienen como denominador común la fatiga, falta de experiencia ante escenarios de emergencia, errores ajenos a la tripulación de la aeronave como la información brindada por la torre de control en tierra y, también, fines malintencionados en muchos casos relacionados a ataques terroristas o algún problema personal psicológico del piloto, el resto, siendo la minoría, obedece a razones técnicas de la aeronave.

Conocidas estas causas que pueden llevar a un accidente aéreo, intentaremos demostrar con datos, que el transporte aéreo es de los más seguros actualmente. Para esto tomaremos un caso específico para comparar con otro medio de transporte en un lugar determinado en un tiempo determinado.
