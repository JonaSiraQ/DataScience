## <a id='3'>3. Matriz de Confusión</a>


Asumamos que el `evento` buscado es que un participante se vaya de la casa, la metodología usada en el programa es por mayor porcentaje de voto `negativo`.

Asumamos que nos posicionamos en quien se fué, plantearemos una MATRIZ DE CONFUSIÓN de la siguiente manera:

Método Indirecto: los sentimientos se traducen en llamadas en contra del opositor
- Participante #1: es `el se fue` 
    - Los sentimientos `positivos` del p#2, se traducen en llamadas para que se vaya el p#1
        - Verdadero Positivo ya que el p#1 se fue
    - Los sentimientos `negativos` del p#2, se traducen en llamadas para que se quede el p#1
        - Falso Negativo ya que el p#1 se fue
                

- Participante #2: es `el se quedó` 
    - Los sentimientos `positivos` del p#1, se traducen en llamadas para que se vaya el p#2
        - Falso Negativo ya que el p#2 se quedó
    - Los sentimientos `negativos` del p#1, se traducen en llamadas para que se quede el p#2
        - Verdadero Negativo ya que el p#2 se quedó
        
Método Directo: los sentimientos se relacionan directamente con lo que pasó posteriormente
- Participante #1: es `el se fue` 
    - Los sentimientos `positivos` del p#1, se traducen en llamadas para que se quede el p#1
        - Falso Negativo ya que el p#1 se fue
    - Los sentimientos `negativos` del p#1, se traducen en llamadas para que se vaya el p#1
        - Verdadero Negativo ya que el p#1 se fue
                

- Participante #2: es `el se quedó`
